Решение:

Пусть X - дискретная случайная величина, равная числу правильных ответов. 
Она распределена по биномиальному закону с параметрами n = 5 (число вопросов), 
p =1/4=0,25 (вероятность угадать ответ), поэтому вероятности вычисляем по формуле Бернулли:
$$P(X = k) = C_n^k p^k (1 - p)^{n - k}$$

Возможные значения такой случайной величины:  x1=0,  x2=1,  x3=2,  x4=3, x5=4, x6=5.

$$P(X = 1) = C_n^k p^k (1 - p)^{n - k} = C_5^1 \cdot 0.25^1 \cdot 0.75^4 = \frac{5!}{1! \: (5-1)!} \cdot 0.25^1 \cdot 0.75^4 = 5 \cdot 0.25 \cdot 0.31640625 = 0.3955078125.$$

$$P(X = 3) = C_n^k p^k (1 - p)^{n - k} = C_5^3 \cdot 0.25^3 \cdot 0.75^2 = \frac{5!}{3! \: (5-3)!} \cdot 0.25^3 \cdot 0.75^2 = 10 \cdot 0.015625 \cdot 0.5625 = 0.087890625.$$

$$P(X = 4) = C_n^k p^k (1 - p)^{n - k} = C_5^4 \cdot 0.25^4 \cdot 0.75^1 = \frac{5!}{4! \: (5-4)!} \cdot 0.25^4 \cdot 0.75^1 = 5 \cdot 0.00390625 \cdot 0.75 = 0.0146484375.$$

$$P(X = 5) = C_n^k p^k (1 - p)^{n - k} = C_5^5 \cdot 0.25^5 \cdot 0.75^0 = \frac{5!}{5! \: (5-5)!} \cdot 0.25^5 \cdot 0.75^0 = 1 \cdot 0.0009765625 \cdot 1 = 0.0009765625.$$

Задача 2

Вероятность попадания в цель при одном выстреле равна 0.01. Какова вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10?

Решение:

Пусть случайная величина X – число попаданий в цель. Так как вероятность $p=0,01$ очень мала, а число выстрелов (опытов) достаточно велико ($n=200$  выстрелов), то искомую вероятность будем находить, используя формулу Пуассона.

$$P(X = k) = \dfrac{a^k e ^{-a}}{k!}.$$

Здесь $a$ — положительное вещественное число.

По теореме сложения вероятностей:

$$P(5<=X<=10) = P(X=5) + P(X=6) + P(X=7) + P(X=8) + P(X=9) + P(X=10).$$

Учитывая, что $a = np$, получим:

$$a = np = 200 * 0,01 = 2.$$.

$$P(5<=X<=10) = \dfrac{a^k e ^{-a}}{k!} = \dfrac{2^5 e ^{-2}}{5!} + \dfrac{2^6 e ^{-2}}{6!} + \dfrac{2^7 e ^{-2}}{7!} + \dfrac{2^8 e ^{-2}}{8!} + \dfrac{2^9 e ^{-2}}{9!} + \dfrac{2^{10} e ^{-2}}{10!} = e ^{-2}(\dfrac{2^5}{5!} + \dfrac{2^6}{6!} + \dfrac{2^7}{7!} + \dfrac{2^8}{8!} + \dfrac{2^9}{9!} + \dfrac{2^{10}}{10!}).$$

$$P(5<=X<=10) = \approx 0.135 (\dfrac{32}{120} + \dfrac{64}{720} + \dfrac{128}{5040} + \dfrac{256}{40320} + \dfrac{512}{362880} + \dfrac{1024}{3628800}) \approx 0.052514286.$$

Задача 3

В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет совершено попадание?

Решение:

Пусть случайная величина X – число попаданий в цель.
Собятия: число попаданий X>=k и число попаданий 0<=X<k образуют полную группу событий.

Поэтому вероятность P(X>=k) будем искать как 1 - P(0<=X<k)

Имеем:

P(X>=10)=1-P(0<=X<10)=1-(P(X=0)+P(X=1)+P(X=2)+P(X=3)+P(X=4)+P(X=5)+P(X=6)+P(X=7)+P(X=8)+P(X=9))

```python
import numpy as np
n=1420
p=0.01
a=n*p
P=0
sum=0
for i in range(0,10):
     sum=sum+(a ** i) * (np.exp(-a)) / np.math.factorial(i)
P=1-sum
print(P)
  
```
